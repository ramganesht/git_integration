```python
# ===============================
# Notebook Initialization
# ===============================

import random
import math
import time

print("Notebook started--::123")

# ===============================
# Utility Functions
# ===============================

def slow_square(x):
    time.sleep(0.0001)
    return x * x

def generate_series(n):
    return [i for i in range(n)]

def noisy_sum(values):
    total = 0
    for v in values:
        total += v + random.randint(-1, 1)
    return total

# ===============================
# Experiment Block 1
# ===============================

results_exp_1 = []

for i in range(100):
    val = slow_square(i)
    results_exp_1.append(val)

print("Experiment 1 done:", len(results_exp_1))

# ===============================
# Experiment Block 2 (Repeated Pattern)
# ===============================

results_exp_2 = []

for i in range(100):
    val = slow_square(i + 1)
    results_exp_2.append(val)

print("Experiment 2 done:", len(results_exp_2))

# ===============================
# Large Loop Section (Diff Stress)
# ===============================

big_results = []

for i in range(5000):
    if i % 500 == 0:
        print("Processing index::", i)
    big_results.append(i * math.sin(i))

print("Big loop finished")

# ===============================
# Data Transformation
# ===============================

transformed = []

for idx, val in enumerate(big_results):
    transformed.append({
        "index": idx,
        "value": val,
        "abs": abs(val),
        "rounded": round(val, 4)
    })

print("Transformation complete------")

# ===============================
# Statistical Summary
# ===============================

values = [item["value"] for item in transformed]

mean_value = sum(values) / len(values)
min_value = min(values)
max_value = max(values)

print("Stats")
print("Mean:", mean_value)
print("Min:", min_value)
print("Max:", max_value)

# ===============================
# Repeated Cells Simulation (Common Notebook Pattern)
# ===============================

def run_experiment(seed):
    random.seed(seed)
    data = []
    for i in range(1000):
        data.append(random.random() * i)
    return {
        "seed": seed,
        "sum": sum(data),
        "max": max(data),
        "min": min(data)
    }

experiment_results = []

for seed in range(10):
    experiment_results.append(run_experiment(seed))

print("Experiments completed:", len(experiment_results))

# ===============================
# Debug / Logging Noise
# ===============================

for i in range(50):
    print(f"[DEBUG] iteration={i}, value={i*i}")

# ===============================
# Intentional Conflict-Friendly Section
# (Great for merge/diff testing)
# ===============================

CONFIG = {
    "timeout": 5000,
    "retries": 3,
    "log_level": "INFO"
}

def process_config(cfg):
    if cfg["timeout"] > 3000:
        print("High timeout")
    else:
        print("Low timeout")

process_config(CONFIG)

# ===============================
# Large Text Output Generator
# ===============================

large_text = []
for i in range(2000):
    large_text.append(f"Line {i}: value={i*i*i}")

print("Generated large text block:", len(large_text))

# ===============================
# Final Cell
# ===============================

print("Notebook execution finished successfully")
```

```python
# ===============================
# Additional Analysis Cell
# ===============================

import statistics

# Reuse previous experiment results if present
if "experiment_results" in globals():
    sums = [r["sum"] for r in experiment_results]
    max_values = [r["max"] for r in experiment_results]

    print("Additional Stats")
    print("Average sum:", statistics.mean(sums))
    print("Median sum:", statistics.median(sums))
    print("Std deviation:", statistics.stdev(sums))
    print("Overall max:", max(max_values))
else:
    print("experiment_results not found")

# ===============================
# Heavy Repetition (Diff / Scroll Stress)
# ===============================

repeated_output = []

for block in range(20):
    block_data = []
    for i in range(500):
        block_data.append({
            "block": block,
            "i": i,
            "value": i ** 2,
            "sqrt": i ** 0.5
        })
    repeated_output.append(block_data)
    print(f"Block {block} generated with {len(block_data)} entries")

print("Repeated blocks complete")

# ===============================
# Intentional Change Hotspot
# (Good for merge conflict demos)
# ===============================

THRESHOLD = 0.75

def filter_values(data, threshold):
    filtered = []
    for item in data:
        if item["value"] % 3 == 0 and item["sqrt"] > threshold:
            filtered.append(item)
    return filtered

sample_block = repeated_output[5]
filtered_sample = filter_values(sample_block, THRESHOLD)

print("Filtered items:", len(filtered_sample))

# ===============================
# Large Print Section (Output Noise)
# ===============================

for i in range(100):
    print(f"[TRACE] cell2 iteration={i}, computed={i * 3.14159}")

print("Second cell execution finished")
```
